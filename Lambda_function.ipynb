```python
# Filter
#syntax:filter(function,iterable)
#find the even
no = [1,2,3,4,5,6]
even = list(filter(lambda x : x%2 ==0,no))
even
```

```python
#syntax:filter(function,iterable)
#find the odd
no = [1,2,3,4,5,6]
odd = list(filter(lambda x : x % 2 !=0,no))
odd
```

```python
#filter strings based on lenght more than 4
words = ["Data","science","AI","ML","Machind"]

string_4 =list(filter (lambda s : len(s)>4,words) )
string_4 
```

```python
#filter strings based on lenght less than 4
words = ["Data","science","AI","ML","Machind"]

string_4 =list(filter (lambda s : len(s)<4,words) )
string_4 
```

```python
#filter negative number from the list
no = [-9,3,-4,5,-8,4]
result = list(filter(lambda x : x<0,no))
result
```

```python
#filter positive number from the list
no = [-9,3,-4,5,-8,4]
result = list(filter(lambda x : x>0,no))
result
```
