```python
# Decorators
def decorators(func):
    def wrappers(*args,**kwargs):
        print("before function execute:")
        result = func(*args,**kwargs)
        print("after function execute:")
        return result
    return wrappers
```

```python
def greet(name):
    print(f"Hello! {name}")
```

```python
greet("R")
```

```python
@ decorators
def greet(name):
    print(f"Hello! {name}")
```

```python
greet("Hanuman")
```

Log Decorator

```python
def log_decorator(func):
    def wrapper(*args,**kwargs):
        print(f"before function execution:{func.__name__}")
        print()
        result = func(*args,**kwargs)
        print()
        print(f"after function execution:{func.__name__}")
        return result
    return wrapper
```

```python
@ log_decorator
def greet(name,age,loc):
    print(f"Hello! {name}: you are {age} years old: you live in {loc}!")
```

```python
greet("Abani",46,"PKL")
```

```python
import time
def decorator_timer(func):
    def wrapper(*args,**kwargs):
        start_time = time.time()
        result = func(*args,**kwargs)
        end_time =time.time()
        exe_time = end_time - start_time
        print(f"{func.__name__} took {exe_time:.4f} seconds to execute")
        return result
    return wrapper()
    
```

```python
@ decorator_timer
def slow_func():
    time.sleep(2)
    return "Done"
```
